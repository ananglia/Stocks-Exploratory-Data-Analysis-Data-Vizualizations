Plot Annual Share growth for the top 2 companies - Google and Amazon

```python
AMZN['year']=[j[0:4] for j in AMZN['Date']]
```

```python
df_AMZN= AMZN.groupby('year')['Close'].mean().to_frame()
```

```python
df_AMZN['Close_lag_yr']=df_AMZN['Close'].shift(-1)
```

```python
df_AMZN.fillna(value=0)
```

```python
GOOGL['year']=[i[0:4] for i in GOOGL['Date']]
```

```python
df_GOOGL = GOOGL.groupby('year')['Close'].mean().to_frame()
```

```python
df_GOOGL['Close_lag_yr']=df_GOOGL['Close'].shift(-1)
```

```python
df_GOOGL.fillna(value=0)
```

```python
df_GOOGL['Annual_growth'] = ((df_GOOGL['Close_lag_yr']-df_GOOGL['Close'])/df_GOOGL['Close'])*100
df_AMZN['Annual_growth'] = ((df_AMZN['Close_lag_yr']-df_AMZN['Close'])/df_AMZN['Close'])*100
```

```python
df_GOOGL.reset_index(inplace=True)
df_AMZN.reset_index(inplace=True)
```

```python
trace = go.Scatter(x = df_GOOGL['year'], y = df_GOOGL['Annual_growth'],
                  name='Annual Google growth rate (in USD)')
trace_1= go.Scatter(x = df_AMZN['year'], y = df_AMZN['Annual_growth'],
                  name='Annual Amazon growth rate (in USD)')

layout = go.Layout(title='Google Vs Amazon Share Annual Growth Rate',
                   plot_bgcolor='rgb(230, 230,230)', 
                   showlegend=True)
fig = go.Figure(data=[trace,trace_1], layout=layout)

py.iplot(fig)
```
